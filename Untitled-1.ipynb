```python
# Import necessary modules from Qiskit
from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
import matplotlib.pyplot as plt

# ========== 1. Create a quantum circuit ==========
# Create a quantum circuit with 2 qubits and 2 classical bits

qc = QuantumCircuit(2, 2)

# ========== 2. Apply quantum gates ==========
# Apply Hadamard gate to create superposition
qc.h(0)

# Apply CNOT gate to create entanglement (Bell state)
qc.cx(0, 1)

# Measure the qubits
qc.measure([0, 1], [0, 1])

# Draw the circuit
print("Quantum Circuit:")
print(qc.draw())

# ========== 3. Simulate with Aer simulator ==========
# Option 1: Basic simulator for measurement results
print("\nOption 1: Basic simulation with QASM simulator")
simulator = AerSimulator()

# Execute the circuit
job = simulator.run(qc, shots=1024)
result = job.result()
counts = result.get_counts(qc)

print("\nMeasurement results:", counts)
print(f"Probability of |00>: {counts.get('00', 0)/1024:.3f}")
print(f"Probability of |11>: {counts.get('11', 0)/1024:.3f}")

# ========== 5. Simulate with noise ==========
print("\n" + "="*50)
print("Option 3: Simulation with noise model")

from qiskit_aer.noise import NoiseModel, depolarizing_error

# Create a noise model
noise_model = NoiseModel()

# Add depolarizing error to single qubit gates
error = depolarizing_error(0.01, 1)  # 1% error rate
noise_model.add_all_qubit_quantum_error(error, ['h'])

# Add depolarizing error to two-qubit gates
error_cx = depolarizing_error(0.02, 2)  # 2% error rate
noise_model.add_all_qubit_quantum_error(error_cx, ['cx'])

# Execute with noise
job_noise = simulator.run(qc, shots=1024, noise_model=noise_model)
result_noise = job_noise.result()
counts_noise = result_noise.get_counts(qc)

print("\nMeasurement results with noise:", counts_noise)

# Compare noisy vs ideal results
ideal_counts = {'00': 512, '11': 512}  # Expected ideal results
print(f"Ideal would be ~50% |00> and 50% |11>")

# ========== 6. Circuit optimization example ==========
print("\n" + "="*50)
print("Circuit optimization example")

# Create a more complex circuit
qc_complex = QuantumCircuit(3, 3)
qc_complex.h(0)
qc_complex.h(1)
qc_complex.cx(0, 2)
qc_complex.cx(1, 2)
qc_complex.measure_all()

print("Original circuit depth:", qc_complex.depth())
print("Original circuit size:", qc_complex.size())

# Transpile with optimization
transpiled_qc = transpile(qc_complex, simulator, optimization_level=3)
print("Transpiled circuit depth:", transpiled_qc.depth())
print("Transpiled circuit size:", transpiled_qc.size())

```

```python
import numpy as np
import matplotlib.pyplot as plt

for si in range(10):
    def f(x):
        # Example periodic function to interpolate
        return fnx(x)
        #return np.cos(3*x) + 0.5*np.sin(5*x) + 0.2 * np.cos(0*x)  # a0 + cos + sin

    # choose R (highest harmonic)
    R = 2
    N = 2*R + 1  # number of sample points

    # sample points
    x_samples = np.linspace(-np.pi, np.pi, N, endpoint=False)
    y_samples,y_p,y_m = f(x_samples)

    # Compute DFT (complex-valued)
    # Use numpy FFT and shift: since data is real and points are symmetric
    Y = np.fft.fft(y_samples)/N  # normalized
    # Frequencies from -R to +R
    freqs = np.fft.fftfreq(N, d=(2*np.pi)/N)  # normalized in cycles per 2π
    # But more simply, we just use k = 0,...,N-1 and treat the symmetry manually

    # Build real trigonometric interpolant p(x)
    def p(x):
        val = Y[0].real  # a0 term 
        diff = Y[0].real*0
        # Positive frequencies
        for k in range(1, R+1):
            ck_plus = Y[k]
            ck_minus = Y[-k % N]  # alias for negative freq
            # Real-valued cosine/sine contribution:
            # a_k = ck_plus + ck_minus
            # b_k = i*(ck_minus - ck_plus)
            a_k = -ck_plus.real - ck_minus.real
            b_k =  (ck_plus.imag - ck_minus.imag)
            val += a_k * np.cos(k * x) + b_k * np.sin(k * x)
            diff += -a_k*k* np.sin(k * x) + b_k *k* np.cos(k * x)
        return val, diff

    # Test reconstruction
    x_plot = np.linspace(-np.pi, np.pi, 500)
    y_true,yt_p,yt_m = f(x_plot)
    y_interp = np.array([p(x) for x in x_plot])

    # Plot
    plt.figure(figsize=(8,4))
    plt.plot(x_plot, y_true, label="Original f(x)")
    plt.plot(x_plot, y_interp, '--', label="Interpolant p(x)")
    plt.plot(x_plot, (np.array(yt_p) - np.array(yt_m))*np.sqrt(0.5), '--', label=r"Interpolant $\partial p(x)$")
    #plt.plot(x_plot, np.array(yt_p), '--', label=r"Interpolant $\partial p(x)$")
    #plt.plot(x_plot, np.array(yt_m), '--', label=r"Interpolant $\partial p(x)$")

    plt.scatter(x_samples, y_samples, color='red', label="Sample points")
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.title(f"Trigonometric interpolation with R = {R}, N = {N}")
    plt.show()
```
